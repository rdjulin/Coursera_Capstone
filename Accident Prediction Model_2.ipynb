```python
limit = 100
df_accidents = df_a2.iloc[0:limit, :]
```

```python
# Seattle latitude and longitude values
latitude = 47.6062
longitude = -122.3321
```

```python
# create map and display it
seattle_map = folium.Map(location=[latitude, longitude], zoom_start=12)
```

```python
# instantiate a feature group for the accidents in the dataframe
accidents = folium.map.FeatureGroup()

# loop through the 100 accidents and add each to the accidents feature group
for lat, lng, in zip(df_accidents.Y, df_accidents.X):
    accidents.add_child(
        folium.features.Marker([lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add incidents to map
seattle_map.add_child(accidents)
```

```python
# create map and display it
seattle_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# loop through the 100 accidents and add each to the map
for lat, lng, label in zip(df_accidents.Y, df_accidents.X, df_accidents.SEVERITYCODE):
    folium.CircleMarker(
        [lat, lng],
        radius=5, # define how big you want the circle markers to be
        color='yellow',
        fill=True,
        popup=label,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(seattle_map)

# show map
seattle_map
```
